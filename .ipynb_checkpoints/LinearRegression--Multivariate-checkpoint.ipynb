```python
X=np.array(df.iloc[:,:5]) #Features(5 features)
Y=np.array(df.iloc[:,5:6]) # Output(Target value)
```

```python
split_value=round(len(X)-len(X)*0.25) #splitting the dataset into 2 parts(75% ,25%)
x_train=X[:split_value] # 75% data is used for training the model
y_train=Y[:split_value]
x_test=X[split_value:] #25% is used for testing the model
y_test=Y[split_value:]
```

# Model for Multivariate Linear Regression

```python
class Regression:
    def __init__(self,x_train):
        n=x_train.shape[0] #using for calculating the value of slope(m)
        self.m=np.zeros((1,n)) # m=slope
        self.c=0 # c=intercept(initially taken as 0)
    
    def hypothesis(self,x_train):
        h=np.dot(self.m,x_train)+self.c # As h=mx+c
        return h    # here m is of 1xn and x_train is of nXm size so H will be of 1xm size
    
    def cost_function(self,h,y_train):
        m=y_train.shape[1]
        cost=((1/(2*m))*np.sum(np.square(h-y_train)))
        return cost
    
    def gradient_descent(self,h,x_train,y_train):
        m=y_train.shape[1]
        dm=(1/m)*np.dot((h-y_train),x_train.transpose()) #derivative of slope(m) for repeateadly updation
        dc=(1/m)*(np.sum(h-y_train)) # dreivative of intercept(c)
        return dm,dc
    
    def training_model(self,x_train,y_train,learning_rate,iterations):
        Cost_list=[]
        Iteration_list=[]
        for i in range(iterations):
            h=self.hypothesis(x_train)
            cost=self.cost_function(h,y_train)
            gradd=self.gradient_descent(h,x_train,y_train)
            m_update=gradd[0]
            c_update=gradd[1]
            self.m=self.m-(learning_rate*m_update)
            self.c=self.c-(learning_rate*c_update)
            if(i%10==0):
                Cost_list.append(cost)
                Iteration_list.append(i)
        plt.plot(Iteration_list,Cost_list)
        plt.xlabel('Iterations')
        plt.ylabel('Cost')
    
    def pred(self,x_test):
        y_pred=np.dot(self.m,x_test)+self.c # predicting the output value(y_pred) for test dataset part
        return y_pred
```

```python
x_train=x_train.transpose() 
y_train=y_train.transpose()
x_test=x_test.transpose()
y_test=y_test.transpose()
learning_rate=0.001
iterations=500
lreg=Regression(x_train) #creating object of regression
lreg.training_model(x_train,y_train,learning_rate,iterations) # training the model for training dataset part
y_pred=lreg.pred(x_test) # prediction of output for test dataset part
mse=np.sum(y_pred-y_test) #calculation of mean square error(mse)
mse
```

# Model from sk_learn

```python
from sklearn.linear_model import LinearRegression
lreg=LinearRegression() #object of linear regression
lreg.fit(x_train.transpose(),y_train.transpose()) #training the model 
y_pred=lreg.predict(x_test.transpose()) # prediction for target value
mse_sklearn= np.sum(y_pred-y_test) #calculation of mean square error(mse)
print(mse_sklearn)
```

# Comparison b/w Scratch & sk_learn

```python
if(mse< mse_sklearn):
    print("Scratch is best")
else:
    print("sk_learn is best")
```
